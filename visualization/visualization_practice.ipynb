```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.set_printoptions(suppress=True)
```

```python
figx, ax = plt.subplots(nrows=1, ncols=1)

def plot_colored_sinusoidal_lines(ax, phase=0):
    """Plot sinusoidal lines with colors following the style color cycle."""
    L = 2 * np.pi
    x = np.linspace(0, L)
    nb_colors = len(plt.rcParams['axes.prop_cycle'])
    shift = np.linspace(0, L, nb_colors, endpoint=False)
    for s in shift:
        ax.plot(x, np.sin(x + s + phase), '-')
    ax.set_xlim([x[0], x[-1]])
    return ax

plot_colored_sinusoidal_lines(ax, phase=0);
```

```python
df = pd.read_csv(
"https://raw.githubusercontent.com/plotly/datasets/master/tips.csv"
)
```

```python
meals = df.total_bill.to_numpy()
tips = df.tip.to_numpy()
per = tips/meals

a = meals - meals.mean()
b = (per - per.mean())

a @ b / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
meal_per = np.stack((meals, per), axis=-1)
meal_per = meal_per[np.argsort(meal_per[:,0])]
```
